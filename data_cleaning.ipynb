```python
chunk_size = 10000  # Adjust the size based on your system's memory
chunks = []
for chunk in pd.read_csv("C:/Users/SAM/Desktop/classification/GUIDE_Test.csv.zip", chunksize=chunk_size):
    chunks.append(chunk)
test = pd.concat(chunks, axis=0)
```

```python
# Get unique values in the column from both datasets
unique_train = set(train['Category'].unique())
unique_test = set(test['Category'].unique())

# Check if test set values are a subset of training set values
is_subset = unique_test.issubset(unique_train)
print(is_subset)  # Should return True if test set has no new unique values
```

```python
train['Timestamp'] = pd.to_datetime(train['Timestamp'], errors='coerce')
test['Timestamp'] = pd.to_datetime(test['Timestamp'], errors='coerce')
```

```python
# Extract year, month, day, and hour from the Timestamp column
train['Year'] = train['Timestamp'].dt.year
train['Month'] = train['Timestamp'].dt.month
train['Day'] = train['Timestamp'].dt.day
train['Hour'] = train['Timestamp'].dt.hour

# Similarly for test_df
test['Year'] = test['Timestamp'].dt.year
test['Month'] = test['Timestamp'].dt.month
test['Day'] = test['Timestamp'].dt.day
test['Hour'] = test['Timestamp'].dt.hour
```

```python
columns_to_drop = [
    'MitreTechniques', 
    'ActionGrouped', 
    'ActionGranular', 
    'EmailClusterId', 
    'ThreatFamily', 
    'ResourceType', 
    'Roles', 
    'AntispamDirection', 
    'SuspicionLevel', 
    'LastVerdict',
    'Timestamp',
     'Unnamed: 0'  
]
```

```python
# Drop columns from the training DataFrame
train.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Drop columns from the testing DataFrame
test.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
# Check remaining columns in train and test DataFrames
print("Remaining columns in training DataFrame:")
print(train.columns)

print("\nRemaining columns in testing DataFrame:")
print(test.columns)
```

```python
# Save the datasets as CSV files
train.to_csv('train_dataset.csv', index=False)
test.to_csv('test_dataset.csv', index=False)
```
